```python
# Write your code to make your dataset analysis ready
# ============================================
# Data Wrangling Code
# ============================================

# 1. Check duplicate rows
print("Duplicate Rows")
print("-" * 40)
print("Titles Dataset :", titles_df.duplicated().sum())
print("Credits Dataset:", credits_df.duplicated().sum())


# 2. Remove duplicate rows
titles_df = titles_df.drop_duplicates()
credits_df = credits_df.drop_duplicates()


# 3. Check missing values
titles_missing = (
    titles_df.isnull().sum() / len(titles_df) * 100
).round(2)

credits_missing = (
    credits_df.isnull().sum() / len(credits_df) * 100
).round(2)

print("\nMissing Values (%) - Titles Dataset")
print(titles_missing[titles_missing > 0].sort_values(ascending=False))

print("\nMissing Values (%) - Credits Dataset")
print(credits_missing[credits_missing > 0].sort_values(ascending=False))


# 4. Merge the datasets
merged_df = pd.merge(
    titles_df,
    credits_df,
    on="id",
    how="left"
)


# 5. Check merged dataset
print("\nMerged Dataset Shape")
print(merged_df.shape)


# 6. Missing values after merge
merged_missing = (
    merged_df.isnull().sum() / len(merged_df) * 100
).round(2)

print("\nMissing Values (%) After Merge")
print(merged_missing[merged_missing > 0].sort_values(ascending=False))
```

```python
# Chart - 6
# Top 10 Genres Distribution

# Convert string representation of lists into actual lists
genres = titles_df['genres'].dropna().explode()

# If genres are stored as strings like "['Drama', 'Comedy']"
genres = (
    titles_df['genres']
    .dropna()
    .str.strip('[]')
    .str.replace("'", "", regex=False)
    .str.split(', ')
    .explode()
)

top_genres = genres.value_counts().head(10)

plt.figure(figsize=(12,6))

ax = sns.barplot(
    x=top_genres.values,
    y=top_genres.index,
    palette='viridis'
)

for container in ax.containers:
    ax.bar_label(container)

plt.title("Top 10 Genres on Amazon Prime", fontsize=16)
plt.xlabel("Number of Titles")
plt.ylabel("Genre")

plt.show()
```
